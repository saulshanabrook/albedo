```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
all_trials = {
    'black': [
        [22, 34, 37, 38, 40, 43, 43, 44, 43, 44, 45, 47, 47],
        [20, 31, 33, 34, 35, 38, 39, 42, 42, 44, 46, 47, 50],
        [28, 40, 41, 43, 44, 46, 48, 48, 50, 52, 54, 55, 57]
    ],
    'white': [
        [22, 30, 31, 33, 35, 36, 38, 39, 40, 41, 42, 44, 45],
        [24, 32, 33, 34, 36, 37, 38, 40, 43, 44, 43, 43, 45],
        [26, 35, 36, 37, 38, 39, 40, 41, 41, 39, 41, 43, 43]
    ]
}
list_of_data = []
for (color, trials) in all_trials.items():
    for trial in trials:
        for (exp_n, temp) in enumerate(trial):
            elapsed_time = exp_n * 5
            list_of_data.append({
                'color': color,
                'time': elapsed_time,
                'temp': temp,
            })
data = pd.DataFrame(list_of_data)
```

First we graph our raw data, with temperature changes over time

```python
sns.lmplot("time", "temp", data, hue="color")
```

now lets try to get a mean of the trials

```python
sns.lmplot("time", "temp", data, hue="color", x_estimator=np.mean)
```

Now lets see what it looks like, if we take a look at change in temperature instead of absolute

```python
last_of_data_diff = []
for (color, trials) in all_trials.items():
    for trial in trials:
        for (exp_n, temp) in enumerate(trial[1:]):
            elapsed_time = (exp_n + 1) * 5
            initial_temp = trial[0]
            last_of_data_diff.append({
                'color': color,
                'time': elapsed_time,
                'change in temp': temp - initial_temp,
            })
data_diff = pd.DataFrame(last_of_data_diff)
```

```python
sns.lmplot("time", "change in temp", data_diff, hue="color")
```

## change in temp per time

now lets try something different. instead of plotting absolute temperature, lets try plotting the derivative, or change from five seconds ago to the next

```python
list_of_data_per_time = []
for (color, trials) in all_trials.items():
    for trial in trials:
        for (exp_n, temp) in enumerate(trial[1:]):
            exp_n += 1 # since we are starting at 1
            elapsed_time = exp_n * 5
            last_temperature = trial[exp_n - 1]
            list_of_data_per_time.append({
                'color': color,
                'time': elapsed_time,
                'change in temp per five seconds': temp - last_temperature,
            })
data_per_time = pd.DataFrame(list_of_data_per_time)
```

```python
sns.lmplot("time", "change in temp per five seconds", data_per_time, hue="color")
```

well that doesn't look very good, cause the first jump is always the largest, lets try removing that

```python
sns.lmplot("time", "change in temp per five seconds", data_per_time[data_per_time.time != 5], hue="color")
```

ok that looks alright, but maybe lets add some mean bars

```python
sns.lmplot("time", "change in temp per five seconds", data_per_time[data_per_time.time != 5], hue="color", x_estimator=np.mean)
```
